```python
features_with_nan=[features for features in mix_dataset.columns if mix_dataset[features].isnull().sum()>1]

for features in features_with_nan:
    print(features,np.round(mix_dataset[features].isnull().mean()*100,4),'% missing values')
```

NUMERIC FEATURES


```python
numeric_features = [features for features in mix_dataset.columns if mix_dataset[features].dtype!='O']
print('No of numerical features is ',len(numeric_features))
mix_dataset[numeric_features].head()
```

NON-NUMERIC FEATURES 

```python
non_numeric_features = [features for features in mix_dataset.columns if mix_dataset[features].dtype=='O']
print('No of non-numerical features is ',len(non_numeric_features))
mix_dataset[non_numeric_features].head()
```

```python
for features in numeric_features:
    data = mix_dataset.copy()
    if 0 in data[features].unique():
        pass
    else:
        data[features]=np.log(data[features])
        data.boxplot(column=features)
        plt.ylabel(features)
        plt.title(features)
        plt.show()
```

filling numeric values with median to avoid outliers 

```python
mix_dataset[numeric_features] = mix_dataset[numeric_features].fillna(mix_dataset[numeric_features].median())
```

```python
for features in non_numeric_features:
    mode_value = mix_dataset[features].mode()
    if not mode_value.empty:
        mix_dataset[features]=mix_dataset[features].fillna(mode_value[0])
        
```

```python
symptom_columns = ['Symptom_1','Symptom_2','Symptom_3','Symptom_4','Symptom_5',
                   'Symptom_6','Symptom_7','Symptom_8','Symptom_9','Symptom_10',
                   'Symptom_11','Symptom_12','Symptom_13','Symptom_14','Symptom_15',
                   'Symptom_16','Symptom_17']

# Get all unique symptoms from all columns
unique_symptoms = set()

for col in symptom_columns:
    unique_symptoms.update(mix_dataset[col].dropna().unique())

# Clean them (remove spaces, lowercase, replace with underscore)
unique_symptoms = sorted([symptom.strip().replace(" ", "_").lower() for symptom in unique_symptoms if isinstance(symptom, str)])

# print(f"Total unique symptoms: {len(unique_symptoms)}")
# print(unique_symptoms[:10])  # Just a preview
```

```python
for symptom in unique_symptoms:
    mix_dataset[symptom] = mix_dataset[symptom_columns].apply(
        lambda row: int(symptom in [str(x).strip().replace(" ", "_").lower() for x in row]),
        axis=1
    )
```

```python
def predict_disease(symptom_text):
    # Normalize user symptoms
    user_symptoms = [s.strip().replace(" ", "_").lower() for s in symptom_text.split(',')]

    # Create input row with 0s
    input_dict = {symptom: 0 for symptom in unique_symptoms}
    for symptom in user_symptoms:
        if symptom in input_dict:
            input_dict[symptom] = 1
    

    # Convert to DataFrame
    import pandas as pd
    input_df = pd.DataFrame([input_dict])[trained_features]

    # Predict
    prediction = model.predict(input_df)[0]
    predicted_disease = Disease_label.inverse_transform([prediction])[0]

    # Get precautions
    precaution_row = mix_dataset[mix_dataset['Disease'] == predicted_disease].iloc[0]
    precautions = ', '.join([
        str(precaution_row['Precaution_1']),
        str(precaution_row['Precaution_2']),
        str(precaution_row['Precaution_3']),
        str(precaution_row['Precaution_4']),
    ])

    return predicted_disease, precautions
```
